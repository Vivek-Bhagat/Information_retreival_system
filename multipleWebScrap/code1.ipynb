```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


def scrape_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    items = soup.find_all('div', class_='piece')
    
    data = []
    for item in items:
        
        title_element = item.find('div', class_='ani-name').find('a')
        title = title_element.text.strip() if title_element else 'N/A'
        sub_count = item.find('span', class_='sub').text.strip() if item.find('span', class_='sub') else 'N/A'
        dub_count = item.find('span', class_='dub').text.strip() if item.find('span', class_='dub') else 'N/A'
        total_count = item.find('span', class_='total').text.strip() if item.find('span', class_='total') else 'N/A'
        
       
        data.append({
            'Anime Title': title,
            'Subbed Episodes': sub_count,
            'Dubbed Episodes': dub_count,
            'Total_Episodes': total_count
        })
    
    return data

# Base URL of the website
base_url = 'https://anix.vc/genre/action?page='
total_pages = 150 

all_data = []
for page_number in range(1, total_pages + 1):
    url = f'{base_url}{page_number}'
    print(f'Scraping page: {url}')
    page_data = scrape_page(url)
    all_data.extend(page_data)


df = pd.DataFrame(all_data)


df.to_csv('scraped_data.csv', index=False)

print('Scraping complete. Data saved to scraped_data.csv')
```
